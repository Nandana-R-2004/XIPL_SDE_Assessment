```python
# ==========================
# Q5 CELL 6
# OpenCLIP Model
# ==========================

import open_clip
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"

model, _, preprocess = open_clip.create_model_and_transforms(
    "ViT-B-32",
    pretrained="laion2b_s34b_b79k"
)

model = model.to(device)
model.eval()

print("✅ OpenCLIP Loaded")
```

```python
from PIL import Image
import numpy as np
from skimage.metrics import structural_similarity

def clip_similarity(img1_path, img2_path):

    img1 = preprocess(Image.open(img1_path)).unsqueeze(0).to(device)
    img2 = preprocess(Image.open(img2_path)).unsqueeze(0).to(device)

    with torch.no_grad():

        feat1 = model.encode_image(img1)
        feat2 = model.encode_image(img2)

        feat1 /= feat1.norm(dim=-1, keepdim=True)
        feat2 /= feat2.norm(dim=-1, keepdim=True)

    return (feat1 @ feat2.T).item()


def identity_score(person_path, output_path):

    person = Image.open(person_path).convert("L").resize((256,256))
    output = Image.open(output_path).convert("L").resize((256,256))

    return structural_similarity(
        np.array(person),
        np.array(output),
        data_range=255
    )


def judge_score(garment,idt):

    overall=(garment+idt)/2

    if overall>0.80:
        score=9
        reason="Excellent try-on quality"

    elif overall>0.70:
        score=8
        reason="Good try-on quality"

    elif overall>0.60:
        score=7
        reason="Acceptable quality"

    else:
        score=6
        reason="Visible artifacts"

    return score,reason

print("✅ Evaluation Loaded")
```

```python
from ultralytics import YOLO

yolo = YOLO("yolov8n.pt")

print("✅ YOLO Loaded")
```

```python
import numpy as np

def check_guardrails(person_img):

    # Detect people
    results = yolo.predict(person_img, verbose=False)

    person_found = False

    for box in results[0].boxes:
        cls = int(box.cls[0])

        # COCO class 0 = person
        if cls == 0:
            person_found = True

            x1, y1, x2, y2 = box.xyxy[0].tolist()

            w = x2 - x1
            h = y2 - y1

            # Very simple pose heuristic
            if w > h:
                return True, "⚠️ Warning: Side pose detected."

            if h / w < 1.4:
                return True, "⚠️ Warning: Person may be seated."

            return True, "✅ Suitable pose."

    return False, "❌ No person detected."
```

```python
import tempfile
import time
from PIL import Image

def run_virtual_tryon(person_img, garment_img):

    start = time.time()

    # Save temporary images
    person_path = tempfile.mktemp(suffix=".png")
    garment_path = tempfile.mktemp(suffix=".png")

    person_img.save(person_path)
    garment_img.save(garment_path)

    # -----------------------------
    # Guardrails
    # -----------------------------
    ok, message = check_guardrails(person_img)

    if not ok:
        return None, message, "", "", ""

    # -----------------------------
    # Generate Mask
    # -----------------------------
    mask_dict = auto_masker(person_img, "upper")

    mask = mask_dict["mask"]

    # -----------------------------
    # Run CatVTON
    # -----------------------------
    result = pipeline(
        image=person_img,
        condition_image=garment_img,
        mask=mask,
        num_inference_steps=30,
        guidance_scale=2.5,
        height=1024,
        width=768,
    )

    # Some versions return a list, others return a PIL image
    if isinstance(result, list):
        result = result[0]

    output_path = tempfile.mktemp(suffix=".png")
    result.save(output_path)

    # -----------------------------
    # Evaluation
    # -----------------------------
    garment = clip_similarity(
        garment_path,
        output_path
    )

    identity = identity_score(
        person_path,
        output_path
    )

    score, reason = judge_score(
        garment,
        identity
    )

    scores = f"""
Garment Fidelity : {garment:.4f}

Identity Preservation : {identity:.4f}

Judge Score : {score}/10

Reason:
{reason}
"""

    processing = f"{time.time()-start:.2f} sec"

    garment_attributes = """
Garment Attributes (Q1)

• Category : Upper Wear
• Color : Auto
• Pattern : Auto
• Sleeve : Auto

(Placeholder for Q1 output)
"""

    return (
        result,
        message,
        garment_attributes,
        scores,
        processing,
    )

print("✅ Main function loaded")
```
